```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
import os
```

```python
PATH="./20181113T220601.csv"
```

```python
orig = pd.read_csv(PATH, delimiter=";\s+", engine='python', header=None, names=["pk","m", "buff", "op", "size", "time"])

```

```python
df = orig.copy()
```

```python
# round sizes into bins
BINS = 100
df['bucket'] = np.floor(df['size'] / (df.m / BINS)) * (df.m / BINS)
```

```python
df
```

```python
df[df.m == 10000]
```

```python
for pk in [10,100, 1000]:
    for m in [1000, 10000, 100000]:
        sub = df[(df.m == m) & (df.pk == pk)]
        #sub = sub[sub.time < 4e9]
        avgs = sub.groupby(['buff', 'bucket', 'op'], as_index=False).mean()
        #sub.sort_values(by='size').plot(x='size', y='time', kind='scatter')
        #isplay(avgs)
        print("M: {0} P-K: {1}P+{1}K".format(m, pk))
        
        avgs['category'] = avgs.buff +" "+ avgs.op
        #display(avgs)
        fg = sb.FacetGrid(data=avgs, hue='category', aspect=2)
        fg.map(plt.plot, 'bucket', 'time').add_legend()
        plt.show()
```
